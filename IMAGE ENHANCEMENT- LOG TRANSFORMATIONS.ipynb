### 1. Spatial Domain:
There is a direct manipulation of a pixel in an image (on the image plane)

![](1.jpg)

### 2. Frequency Domain:
Processing the image is based on modifying the Fourier transform of an image.

![](2.jpg)

# Why is it done?

# LOG TRANSFORMATIONS:


#### Properties:

1. For lower amplitudes of input image the range of grey levels is expanded.

2. For higher amplitudes of an input image, the range of grey levels is compressed.

![](log.png)

LOG transform enhances small magnitude input values into wider range of output pixel values and compresses large magnitude input values into narrow range of output values.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
im = cv2.imread('cam.jpg',0)
```

```python
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(121)
plt.title('Original',fontsize=18)
plt.imshow(im,cmap='gray')
```

```python
im_log_law_transformation = 0.6 * (np.log(1 + np.float32(im)))
```

```python
ax = fig.add_subplot(122)
plt.imshow(im_log_law_transformation,cmap='gray')
ax.set_title('log Transform',fontsize=18)
```

# All in one

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

im = cv2.imread('cam.jpg',0)
im_log_law_transformation = 0.6 * (np.log(1 + np.float32(im)))

fig = plt.figure(figsize=(15,5))

ax = fig.add_subplot(121)
plt.title('Original',fontsize=18)
plt.imshow(im,cmap='gray')

ax = fig.add_subplot(122)
plt.imshow(im_log_law_transformation,cmap='gray')
ax.set_title('log Transform',fontsize=18)
```
